```python
data['age'] = pd.qcut(data["age"], q=3, labels=['faixa-1', 'faixa-2', 'faixa-3'])
```

```python
data['health'] = pd.qcut(data["health"].rank(method='first'), q=5, labels=['health-1', 'health-2', 'health-3', 'health-4', 'health-5'])
```

```python
data['absences'] = pd.qcut(data["absences"].rank(method='first'), q=5
                           ,labels=['absences-1', 'absences-2', 'absences-3', 'absences-4', 'absences-5'])
```

```python
data['G3'] = pd.qcut(data["G3"], q=4
                           ,labels=['G3-1', 'G3-2', 'G3-3', 'G3-4'])
```

```python
data['studytime'] = pd.qcut(data["studytime"].rank(method='first'), q=4
                           ,labels=['studytime-1', 'studytime-2', 'studytime-3', 'studytime-4'])
```

```python
rows = data.shape[0]
```

```python
atributos = set()
for i in range(rows):
    atributos = atributos.union(set(data.iloc[i].unique()))
```

```python
atributos
```

```python
np.nan in atributos
```

```python
atributos.difference_update({np.nan})
```

```python
data_ = pd.DataFrame(columns=atributos,data=np.zeros((rows,len(atributos))))
```

```python
data_
```

```python
for i in range(rows):
    data_.at[i, data.iloc[i].dropna()] = 1.
```

```python
data_
```

```python
def supp(df_,X):
    union = np.prod(df_[X].values,axis=1)
    return len(np.nonzero(union)[0])/df_.shape[0]
```

```python
def confidence(df_, X, Y):
    return supp(df_,X+Y)/supp(df_,X)
```

```python
def lift(df_, X, Y):
    return supp(df_,X+Y)/(supp(df_,X)*supp(df_,Y))
```

```python
frequent_itemsets = apriori(data_, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
frequent_itemsets = apriori(data_, min_support=0.05, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
frequent_itemsets = apriori(data_, min_support=0.1, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
rules.sort_values(by='confidence', ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules.sort_values(by='confidence', ascending=False)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
rules.sort_values(by='confidence', ascending=False)
```
